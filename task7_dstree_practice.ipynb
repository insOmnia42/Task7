##### Задача 1. Построение "среднего" алгоритма
В этом задании вам нужно построить графики, демонстрирующие, как алгоритм аппроксимирует истинную зависимость в данных и как он меняется в зависимости от гиперпараметров метода обучения.

````python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
%matplotlib inline
````

````python
def f(x):
    return np.sin(x)   # истинная зависимость в данных
sample_size = 100      # длина выборки
samples_num = 20       # количество выборок
linspace = np.linspace(0, 7, 1000)  # точки для построения графиков
````

__1. (2 балла)__
1. Сгенерируйте выборку $x$ из одномерного экспоненциального распределения (np.random.exponential) длины sample_size.
1. Создайте вектор целевых переменных $y$ как сумму $f(x)$ и случайного шума, сгенерированного из равномерного распределения на отрезке $[-1, 1]$ (np.random.uniform).
1. Обучите DecisionTreeRegressor с параметрами по умолчанию на полученной выборке и сделайте предсказания для объектов из linspace.
1. Постройте два графика на одном рисунке: $f(x)$ и зависимость, восстановленную решающим деревом.

    Рекомендация: не забудьте, что все методы обучения в sklearn требуют на вход двумерную матрицу объекты-признаки. Сделать такую из одномерного вектора можно добавлением мнимых осей (np.newaxis).

````python
np.random.seed(42)
x = np.random.exponential(scale=1.0, size=sample_size)
x = np.sort(x)
noise = np.random.uniform(-1, 1, sample_size)

y = f(x) + noise
x = x[:, np.newaxis]
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(x, y)

x_pred = linspace[:, np.newaxis]
y_pred = regressor.predict(x_pred)

plt.figure(figsize=(10, 6))
plt.plot(linspace, f(linspace), label="Истинная зависимость f(x)", color="blue", linewidth=2)
plt.scatter(x, y, label="Выборка с шумом", color="green", alpha=0.5)
plt.plot(linspace, y_pred, label="Решащее дерево", color="red", linewidth=2)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Аппроксимация зависимости f(x) решающим деревом")
plt.grid()
plt.show()
````

__2. (1 балл)__

Повторите первые 3 шага, описанные выше, samples_num раз. На одном графике для каждого обученного решающего дерева визуализируйте восстановленную им зависимость (рекомендуется все такие линии рисовать полупрозрачными и серым цветом: plt.plot(...... color="gray", alpha=0.5)).  На этом же графике изобразите истинную зависимость f(x) (красным цветом: color="red") и усредненную по всем деревьям восстановленную зависимость (черным цветом: color="black").    

````python
def depth(max_depth=None):
    plt.figure(figsize=(12, 8))

    all_predictions = []

    for i in range(samples_num):
        x = np.random.exponential(scale=1.0, size=sample_size)
        x = np.sort(x)
        noise = np.random.uniform(-1, 1, sample_size)
        y = f(x) + noise
        x = x[:, np.newaxis]

        regressor = DecisionTreeRegressor(max_depth=max_depth, random_state=i)
        regressor.fit(x, y)

        y_pred = regressor.predict(x_pred)
        all_predictions.append(y_pred)
        plt.plot(linspace, y_pred, color="gray", alpha=0.5)

    avg_prediction = np.mean(all_predictions, axis=0)
    plt.plot(linspace, f(linspace), label="Истинная зависимость f(x)", color="red", linewidth=2)
    plt.plot(linspace, avg_prediction, label="Усреднённое предсказание", color="black", linewidth=2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Глубина дерева: {max_depth}")
    plt.legend()
    plt.grid()
    plt.show()

depth(None)
````
